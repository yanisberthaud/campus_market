# Cahier des Charges - Campus Market 


## 1. Présentation du Projet

- **Nom** : Campus Market
- **Concept** : Plateforme web sécurisée facilitant l'achat, la vente et l'échange de matériel entre étudiants.
- **Public Cible** : Étudiants et personnel administratifs du campus. 


## 2. Objectifs Techniques
- **Environnement de Développement** : VS Code / Serveur Local (XAMPP).
- **Langage** : Php en Programmation Orientée Objet (POO).
- **Base de Données** : MySQL avec interface PDO pour la sécurité.
- **Interface** : HTML, CSS, JavaScript (Front-End).
- **Versionnage** : Git avec un historique de commits propre.

## 3. Spécifications Fonctionnelles

### A. Gestion des utilisateurs
- **Inscription** : Création de compte avec nom, prénom, email étudiant et mot de passe sécurisé (password_hash). 
- **Connexion** : Espace sécurisé par sessions PHP.
- **Profil** : Consultation et modification des informations personnelles par l'utilisateur. 
- **Rôles** : Système de permissions distinguant les "Utilisateurs" des "Administrateurs". 

### B. Gestion des Annonces 
- **Consultation** : Page d'accueil listant les annonces (titre + image) et page de détail complète. 
- **Actions (CRUD) :** 
    - **Créer** : Formulaire d'ajout (titre, description, prix, catégorie, upload d'image).
    - **Lire** : Affichage public des annonces. 
    - **Modifier/Supprimer** : Actions réservées exclusivement à l'auteur de l'annonce ou à l'administrateur.
- **Recherche & Navigation** : 
    - Recherche par mots-clés.
    - Filtres par catégorie(Informatique, Livres, Fournitures, etc.).
    - Tri par prix ou par date. 
    - Pagination des résultats.

### C. Back-Office Administrateur
- **Modération** : Possibilité de supprimer n'importe quelle annonce non conforme. 
- **Gestion des membres** : Liste des utilisateurs, modification des droits ou suspension de compte.

## 4. Architecture de la Base de Données
- **Users** : id, nom, prenom, email, password, role, created_at.
- **Annonces** : id, titre, description, prix, image_path, category_id, user_id, created_at.
- **Categories** : id, libelle.

## 5. Architecture Technique (MVC & Sécurité)

### Structure des dossiers (VS Code) : 
- /src/Models/ : Classes PHP gérant les accès à la base de données via PDO. 
- /src/Controllers/ : Logique de traitement des requêtes.
- /src/Views/ : Fichier **.php** contenant uniquement du HTML et de l'affichage. 
- /public/ : Point d'entrée **index.php**, fichiers CSS, JS et images uploadées. 
- /config/ : connexion à la base de données.


## 6. Contraintes de Sécurité
- **Injections SQL** : Utilisation systématique de requêtes préparées. 
- **Failles XSS** : Échappement des données à l'affichage.
- **Accès** : Vérification systèmatique des droits avant d'autoriser une action de modification ou de suppression.

## 7. Stratégie de Déploiement (Dernière Séance)

- **Hébergement** : Déploiement sur VPS (Linux). 
- **Serveur Web** : Nginx ou Apache. 
- **Versionning** : Git avec commits réguliers et historique clair. 

## 8. Organisation du Travail (Jalons)
- **Séance 1** : Cadrage, Cahier des charges et initialisation Git/MVC. 
- **Séance 2** : Création de la BDD et squelette de l'application (Connexion BDD).
- **Séance 3** : Système d'authentification et gestion des rôles. 
- **Séance 4** : CRUD complet des annonces, filtres et recherche. 
- **Séance 5** : Déploiement sur VPS, tests finaux et remise du dossier.
