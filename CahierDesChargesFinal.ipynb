# Cahier des Charges Final - The Blue Devil Exchange 

## 1. Rappel du Cocnept & Public Cible.

- **Nom Final** : The Blue Devil Exchange
- **Concept** : Plateforme communautaire sécurisée de vente et d'échange de matériel entre étudiants. 
- **Cible** : Etudiants et personnel de l'écosytème de l'université de Duke (North Carolina). 


## 2. Réalisation Technique (Bilan)

- **Environnement** : Développé sous VS Code/XAMPP, déployé avec succès sur un environnement mutualisé LAMP (InfinityFree). 
- **Architecture** : Respect du pattern MVC pour une séparation nette de la logique et de l'affichage. 
- **Sécurité appliquée** : 
    - Utilisation systèmatique de PDO et de requêtes préparées contre les injections SQL. 
    - Hashage des mots de passe avec password_hash(). 
    - Contrainte d'unicité sur les emails et gestion des exceptions SQL. 
    - Sécurisation du répertoire racine via un fichier .htaccess. 

## 3. Spécifications Fonctionnelles Délivrées

- A. Espace Membre : 
    - Inscription/Connexion : Flux complet avec persistance de session via **session_start()**.
    - **Expérience "Sans Friction"** : Connexion automatique après inscription pour améliorer l'engagement utilisateur. 
    - **Dashboard Personnel** : Interface **my_listings.php** permettant la gestion (viualisation, modification, suppression) de son propre inventaire. 

- B. Le MarketPlace (CRUD Complet) : 
    - **Consultation Dynamique** : Galerie d'offres (**browse.php**) tirant les données en temps réel de la table **listings**.
    - **Recherche Globale** : Barre de recherche intégrée au **header.php** utilisant l'opérateur SQL **LIKE**.
    - **Système de Tri & Filtrage** : Tri par prix/date et filtrage dynamique par catégories (Textbooks, Electronics, etc). 
    - **Gestion des Médias** : Upload d'images avec génération de noms de fichiers uniques pour éviter les collisions sur le serveur. 

## 4. Architecture de la Base de Données (Production)

- **Tables Finales** : 
    - **Users** : Gestion des identités et des rôles. 
    - **Listings** : Stockage des annonces avec liaison **user_id** et **category_id**, incluant les chemins d'images (image_url) et le statut. 
    - **Categories** : Référentiel des catégories d'objets. 

## 5. Stratégie de Déploiement & Maintenance

- **Serveur de Production** : Migration vers l'infrastructure **InfinitiFree** via le protocole FTP (FileZilla). 
- **Maintenance** : Centralisation des configurations de connexions dans **Config/Database.php** pour faciliter les bascules environnementales.
